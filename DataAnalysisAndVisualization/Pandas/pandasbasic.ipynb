## Pandas
* Pandas is a library used for data manipulation and analysis in Python.
* It has functions for analyzing, cleaning, exploring, and visualizing data.
* Pandas allow us to analize big data sets easily and efficiently.

````python
import pandas as pd

mydataset = {
  'cars': ["Ram", "Shyam", "Bibek"],
  'Marks': [30, 70, 20]
}

myvar = pd.DataFrame(mydataset)

print(myvar)
````

````python
# Series 
# pandas series are like column in a table.
# it is 1d aray holding data of any type.

a = [1, 7, 2]
myvar = pd.Series(a)
print(myvar)
print(myvar[1])
````

````python
# index argument is used to name the indexes.
a = [1, 7, 2]
myvar = pd.Series(a, index = ["x", "y", "z"])
print(myvar)
````

````python
print(myvar["y"])
````

````python
# can alsoo be made from dictionary
calories = {"ram": 420, "shyam": 380, "deepak": 390}
myvar = pd.Series(calories)
print(myvar)
````

````python
# Data frames 
# Data frame is a 2d data structure, like a table with rows and columns.
data = {
  "length": [420, 380, 390],
  "time": [5, 4, 4.5]
}
myvar = pd.DataFrame(data)
print(myvar)
````

````python
data = {
    "Subject": ['Math', 'Science', 'English'],
    "Marks": [85, 90, 78]
}
myvar = pd.DataFrame(data, index=["Ram", "Shyam", "Bibek"])
print(myvar)
````

````python
## cleaning data
df = pd.read_csv('clean.csv')
#
df['Date'] = pd.to_datetime(df['Date'], format='mixed')# to convert the date column to datetime format
print(df)
````

````python
df.dropna(subset=['Date'], inplace=True) # to remove rows with NaN in Date column
print(df)
````

````python
df.loc[7, 'Duration'] = 250.0 # to fill specific value
print(df)
````

````python
for x in df.index:
  if df.loc[x, "Duration"] > 120:
    df.loc[x, "Duration"] = 120
print(df)
````

````python
##duplicate data on 11 and 12
print(df.duplicated())
````

````python
# removing duplicate
df.drop_duplicates(inplace=True)
print(df)
````

````python
# pandas plotting
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('clean.csv')
df.plot()
plt.show()
````

````python
#scatter plot
df.plot(kind = 'scatter', x = 'Duration', y = 'Calories')
plt.show()
````
