## Scatter plot
The scatter() function plots one dot for each observation in the dataset, using the specified marker style. it need two array of the same length: one for x coordinates and one for y coordinates.

```python
import numpy as np
from numpy import random
x = random.randint(20, size=20) 
y = random.randint(20, size=20)
plt.scatter(x, y)

print(x)
print(y)
```

### Size
You can change the size of the dots with the s argument.

Just like colors, make sure the array for sizes has the same length as the arrays for the x- and y-axis:

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show()
```

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.3) #alpha-> transparency

plt.show()
```

```python
### color size scatter all in one

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()# show color scale

plt.show()
```

## Bar


# bar() function
The bar() function creates a bar chart. You need to provide the positions of the bars on the x-axis and the heights of the bars.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
#for horizontal bar chart use plt.barh(x,y)
plt.barh(x,y)
plt.show()
```

```python
plt.bar(x,y , color = "red", width=0.5)
plt.show()
plt.barh(x,y , color = "green", height=0.5)
plt.show()
```

### Histogram
A histogram is used to show the frequency distribution of a set of continuous or discrete data.
* hist() function is used to create histograms in Matplotlib. 

```python
data = np.random.randn(1000)
plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
print(data)
```

```python

data1 = np.random.randn(1000)
data2 = np.random.randn(1000) + 2  

plt.hist([data1, data2], bins=30, color=['blue', 'orange'], label=['Data1', 'Data2'], alpha=0.7)
plt.title("Multiple Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
## pie chart
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels)
plt.show()
```
